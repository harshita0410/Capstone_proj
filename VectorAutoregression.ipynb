# Alberta

```python
import pandas as pd
from statsmodels.tsa.api import VAR
from statsmodels.tsa.stattools import adfuller

# Assuming 'ConstructionBreton' is your DataFrame containing the data
ConstructionBreton = pd.read_csv('Final.csv')
Construction_Breton = ConstructionBreton[ConstructionBreton['Station Name'] == 'Breton']
Construction_Breton.set_index('Date', inplace=True)

# Selecting columns to include in the analysis, excluding 'provincename', 'Population', 'Station Name', 'Longitude (x)', and 'Latitude (y)'
variables = Construction_Breton.drop(columns=['provincename', 'Population','Transportation','Agriculture', 'Station Name', 'Longitude (x)', 'Latitude (y)','Heat Deg Days (°C)','Cool Deg Days (°C)','Min Temp (°C)','Max Temp (°C)'])

# Check for stationarity using Augmented Dickey-Fuller test
for column in variables.columns:
    result = adfuller(variables[column])
    print(f'ADF Statistic for {column}: {result[0]}')
    print(f'p-value: {result[1]}')

# Define lag order for VAR model
lag_order = 2  # Adjust as needed

# Fit VAR model
model = VAR(variables)
results = model.fit(lag_order)

# Print summary of the model
print(results.summary())

# Forecast future values
forecast = results.forecast(variables.values[-lag_order:], steps=5)  # Adjust 'steps' for desired forecast horizon
print(forecast)
```
